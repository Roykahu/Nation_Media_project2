```python
#imitating Jeonju/Noske and Jeonju/Norske in the newsprint column since they are all the same
concatenated_df['NEWSPRINT (MAKE)'] = concatenated_df['NEWSPRINT (MAKE)'].replace('Jeonju/Noske', 'Jeonju/Norske')
concatenated_df['NEWSPRINT (MAKE)'] = concatenated_df['NEWSPRINT (MAKE)'].replace('Reso/Norske', 'Norske/Resolute')
```

```python
concatenated_df.info()
```

```python
#Filling missing values by joining similar columns
concatenated_df['combined_copies_delivered'] = concatenated_df['Copies Delivered'].fillna(concatenated_df['COPIES DELIVERED'])
concatenated_df.drop(['Copies Delivered', 'COPIES DELIVERED'], axis=1, inplace=True)
```

```python
concatenated_df['combined_print_order'] = concatenated_df['Print Order'].fillna(concatenated_df['PRINT ORDER'])
concatenated_df.drop(['Print Order', 'PRINT ORDER'], axis=1, inplace=True)
```

```python
concatenated_df.info()
```

```python
#Converting the date column into days of the week

# Step 1: Convert the 'Date' column to the datetime format
concatenated_df['Date'] = pd.to_datetime(concatenated_df['Date'], format='%d-%m-%Y')

# Step 2: Convert the date to the desired format (e.g., "Monday", "Tuesday", etc.)
concatenated_df['Day_of_Week'] = concatenated_df['Date'].dt.strftime('%A')

# Display the updated DataFrame
concatenated_df.head()

```
